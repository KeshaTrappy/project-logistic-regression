Также отметим, что такие признаки как `last_coffee_type`, `preferred_roast`, `milk_preference`, `coffee_bean_origin`, `last_drink_size`, `coffee_preference_change` описывают индивидуальные вкусовые предпочтения человека и едва ли могут хоть как-то показывать сигналы к уходу клиента. Убедимся, что они не кореллируют с целевой переменной. Для этого построим матрицу корелляции.

```python
# Проверяем корреляцию между признаками и целевой переменной
corr_matrix = df[['last_coffee_type',
                  'preferred_roast',
                  'milk_preference',
                  'coffee_bean_origin',
                  'last_drink_size',
                  'coffee_preference_change',
                  'churn']].phik_matrix(interval_cols=['coffee_preference_change'])

#Оставляем только столбец с целевой переменной
corr_matrix = corr_matrix.loc[corr_matrix.index != 'churn'][['churn']].sort_values(by='churn', ascending=False)

# Визуализируем результат с помощью тепловой карты
plt.figure(figsize=(3, 5))
sns.heatmap(data=corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu', linewidths=0.5)
plt.title("Матрица корреляции между признаками \n и целевой переменной")
plt.xlabel('Целевая переменная')
plt.ylabel('Признаки')
plt.show()
```

Как видно кореляция данных признаков с целевой переменной практически нулевая.

Остальные признаки так или иначе могут коррелировать с целевой переменной, однако силу их влияния можно оценить только после построения матрицы корреляции. На данном этапе с точки зрения бизнес-логики попробуем разбить признаки на 3 группы - высокая, средняя и низкая мера влияния на целевую переменную.

**Высокая мера влияния**:

- `days_since_last_order`: чем дольше клиент не заказывает, тем выше вероятность, что он уже ушёл;
- `order_frequency_month` и `order_frequency_week`: cнижение частоты заказов - сигнал потери интереса;
- `total_spent_last_week` и `total_spent_last_month`: резкое падение трат в текущем месяце в сравнении с предыдущим говорит о возможности ухода;
- `review_rating_last_1`: низкая оценка последнего заказа - сильный индикатор недовольства и как следствие ухода;
- `subscription_status`: лояльные пользователи, пользующиеся подпиской, менее подвержены уходу;
- `app_opens_per_week`: cнижение активности в прилложении может свидетельствовать о скором уходе.

**Средняя мера влияния**:

- `discount_usage_rate`: клиенты, пользующиеся сервисом только в период скидок, при их отсутствии могут уйти, т.е. высокая доля скидок может означать низкую лояльность;
- `days_since_last_promo`: долгая реакция на промокоды и акции - сигнал потери интереса;

```python
# Создадим функцию для отрображения количества пропусков в абсолютных и относительных значениях
def show_missing_stats(my_df):
    # Датафрейм с кол-вом пропусков по каждому столбцу
    missing_stats = pd.DataFrame({
        'Кол-во пропусков': my_df.isnull().sum(),
        'Доля пропусков': my_df.isnull().mean()
    })
    missing_stats = missing_stats.loc[missing_stats['Кол-во пропусков'] > 0]
    
    if missing_stats.empty:
        return "Пропусков в данных нет"
    
    # Форматируем при выводе через Styler
    return missing_stats.sort_values(by='Кол-во пропусков', ascending=False).style.format({'Доля пропусков': '{:.2%}'})

# Выполняем функцию
show_missing_stats(df)
```

Построим гистограммы и диаграммы размаха для визуализации распределения значений в числовых столбцах.

```python
def show_numeric_hists_and_box(my_df):
    # Записываем названия числовых столбцов
    cols = my_df.select_dtypes(include=['number']).columns.to_list()
    
    # Исключаем целевую переменную
    cols = cols[:-1]

    n_vars = len(cols)
    n_cols = 2
    n_rows = (n_vars + n_cols - 1) // n_cols

    # Создаем фигуру (гистограмма + боксплот)
    fig, axes = plt.subplots(
        nrows=n_rows * 2, 
        ncols=n_cols, 
        figsize=(6 * n_cols, 5 * n_rows),
        gridspec_kw={"height_ratios": [0.8, 0.2] * n_rows},
        sharex=False
    )

    # Превращаем в плоский список для удобства итерации
    axes = axes.flatten() if n_vars > 1 else axes

    for i, col in enumerate(cols):
        # Вычисляем индексы для пары графиков
        idx_hist = (i // n_cols) * (n_cols * 2) + (i % n_cols)
        idx_box = idx_hist + n_cols
        
        ax_hist = axes[idx_hist]
        ax_box = axes[idx_box]

        mean_val = my_df[col].mean()
        median_val = my_df[col].median()
        p99 = my_df[col].quantile(0.99)

        # Строим гистограмму
        sns.histplot(data=my_df, x=col, ax=ax_hist, bins=30, edgecolor='black', kde=False)
        ax_hist.axvline(median_val, color='green', linestyle='--', linewidth=1.5, label=f'Медиана: {median_val:.1f}')
        ax_hist.axvline(mean_val, color='orange', linestyle='--', linewidth=1.5, label=f'Среднее: {mean_val:.1f}')
        ax_hist.axvline(p99, color='red', linestyle='--', linewidth=1.5, label=f'99-й процентиль: {p99:.1f}')
        
        ax_hist.set_title(f'Распределение {col}')
        ax_hist.set_ylabel('Частота')
        ax_hist.legend()

        # Строим диаграмму размаха
        sns.boxplot(data=my_df, x=col, ax=ax_box, fliersize=3)
        ax_box.axvline(mean_val, color='orange', linestyle='--', linewidth=1.5)
        ax_box.axvline(p99, color='red', linestyle='--', linewidth=1.5)
        
        # Синхронизируем оси X
        ax_box.set_xlim(ax_hist.get_xlim())
        ax_box.set_xlabel(col)

    # Скрываем неиспользуемые ячейки
    for j in range(n_vars, n_rows * n_cols):
        idx_empty_hist = (j // n_cols) * (n_cols * 2) + (j % n_cols)
        idx_empty_box = idx_empty_hist + n_cols
        axes[idx_empty_hist].set_visible(False)
        axes[idx_empty_box].set_visible(False)

    plt.tight_layout()
    plt.show()

# Пример вызова:
show_numeric_hists_and_box(df)
```

```python
# Категориальные признаки для кодирования OneHotEncoder и заглушки no_info
cat_oeh_cols = ['subscription_status',
                'coffee_bean_origin',
                'seasons',
                'milk_preference',
                'last_drink_size',
                'phone_type',
                'last_coffee_type',
                'preferred_roast']

# Категориальные признаки для кодирования TargetEncoder и заглушки no_info
cat_te_cols = ['geo_location']

# Числовые признаки для заполения нулем
num_zero_cols = ['seasonal_menu_tried',
                 'coffee_preference_change',
                 'notifications_enabled']

# Числовые признаки для заполнения медианой
num_med_cols = ['days_since_last_order',
                'total_spent_last_week',
                'app_opens_per_week',
                'total_spent_last_month',
                'app_crashes_last_month',
                'discount_usage_rate',
                'review_rating_last_1',
                'median_order_value',
                'review_rating_last_10',
                'order_frequency_month',
                'avg_order_value',
                'order_frequency_week']

# Числовые признаки для биннинга
num_bin_cols = ['days_since_last_promo']
```

Создадим функцию `bin_days_since_last_promo` для биннинга числового столбца 

```python
# Функция для биннинга столбца
def bin_days_since_last_promo(X):
    bins = [-np.inf, 0, 7, 30, np.inf]
    labels = ['today', '1-7_days', '8-30_days', '30+_days']
    X_binned = pd.cut(X.iloc[:,0], bins=bins, labels=labels).astype('object')
    return X_binned.fillna('never_used').values.reshape(-1, 1)
    
# Оборачиваем функцию в трансформер
binner = FunctionTransformer(bin_days_since_last_promo,
                             validate=False,
                             feature_names_out='one-to-one')
```

Создадим пайплайн для обработки данных.

```python
# Заполнение пропусков и кодировка категориальных признаков OneHotEncoder
cat_ohe_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='no_info')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Заполнение пропусков и кодировка категориальных признаков TargetEncoder
cat_te_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='no_info')),
    ('encoder', TargetEncoder())
])

# Биннинг и кодирование OneHotEncoder признака days_since_last_promo
num_bin_pipeline = Pipeline([
    ('binner', binner),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Пайплайн предобработки
preprocessor = ColumnTransformer(
    transformers=[
        ('cat_ohe', cat_ohe_pipeline, cat_oeh_cols),
        ('cat_te', cat_te_pipeline, cat_te_cols),
        ('num_zero', SimpleImputer(strategy='constant', fill_value=0), num_zero_cols),
        ('num_med', SimpleImputer(strategy='median'), num_med_cols),
        ('num_bin', num_bin_pipeline, num_bin_cols)
    ]).set_output(transform='pandas')
```

В качестве проверки выполним предобработку и выведем информацию о преобразованном датафрейме.

```python
# Выполняем трансформацию
X_transformed = preprocessor.fit_transform(X_train_val, y_train_val)

# Выводим информацию
X_transformed.info()
```

- `rating_drop_idx` = `review_rating_last_1` / `review_rating_last_10` (ситуация "все было хорошо, но последний раз меня ужасно обслужили")
- `share_crashes` = `app_crashes_last_month / 4` × `num_zero__app_opens_per_week` (доля вылетов приложения)

Создадим класс для генерации новых признаков `FeatureEngineer`.

```python
# Создаем новый класс для генерации новых признаков
class FeatureEngineer(BaseEstimator, TransformerMixin):
    
    # Инициализируем
    def __init__(self, drop_original=True):
        self.drop_original = drop_original
        # Список исходных колонок, которые мы заменим 
        self.to_replace = [
            'num_med__total_spent_last_month',
            'num_med__app_opens_per_week',
            'num_med__app_crashes_last_month',
            'num_med__review_rating_last_1',
            'num_med__review_rating_last_10']
    
    # Трансформер не обучается поэтому fit должен вернуть self  
    def fit(self, X, y=None):
        return self
    
    # Трансформация признаков
    def transform(self, X):
        X = X.copy()
        
        # Логарифмирование и взятие корня признаков
        X['feat__total_spent_last_month_sqrt'] = np.sqrt(X['num_med__total_spent_last_month'])
        X['feat__app_opens_per_week_sqrt'] = np.sqrt(X['num_med__app_opens_per_week'])
        X['feat__app_crashes_last_month_squared'] = X['num_med__app_crashes_last_month'] ** 2
        
            
        # Создание признака rating_drop_idx
        X['feat__rating_drop_idx'] = (
            X['num_med__review_rating_last_1'] / (X['num_med__review_rating_last_10'] + 0.1))
            
        # Создание признака share_crashes
        X['feat__share_crashes'] = (
            X['num_med__app_crashes_last_month'] * 0.25) / (X['num_med__app_opens_per_week'] + 0.1)
        
        if self.drop_original:
            X = X.drop(columns=self.to_replace)
        return X
    
    
    def get_feature_names_out(self, input_features=None):
        if input_features is None:
            return None
        
        # Список новых колонок
        new_features = [
            'feat__total_spent_last_month_sqrt',
            'feat__app_opens_per_week_sqrt',
            'feat__app_crashes_last_month_squared',
            'feat__rating_drop_idx',
            'feat__share_crashes']
        
        # Формируем список оставшихся колонок
        if self.drop_original:
            current_features = [f for f in input_features if f not in self.to_replace]
        else:
            current_features = list(input_features)
            
        return np.array(current_features + new_features)
```

Создадим новый пайплайн предобработки данных, который включает в себя конструирование новых признаков.

```python
# Полная предобработка с новыми признаками
new_preprocessor = Pipeline([
    # Базовый препроцессор
    ('base_prep', preprocessor),
    # Генератор новых признаков
    ('feature_engineering', FeatureEngineer(drop_original=True)),
    # Стандартизация признаков
    ('scaler', StandardScaler())
]).set_output(transform='pandas')
```

Создадим класс `FeatureSelector`, который оставит только необходимые признаки.

```python
class FeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, feature_names):
        self.feature_names = feature_names
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X[self.feature_names]
    
    def get_feature_names_out(self, input_features=None):
        return self.feature_names
```

**1.** Перечислите все гиперпараметры, с которыми планируете экспериментировать.

Перечислим параметры, которые будем варьировать.
- Сила регуляризации `C`: 0.001, 0.01, 0.1, 1, 10, 100
- Тип штрафа регуляризации `penalty`: 'l1', 'l2', None
- Алгоритм оптимизации `solver`: 'lbfgs', 'liblinear'

**2.** Проведите систематический перебор гиперпараметров для `LogisticRegression`, выполните кросс-валидацию для каждой конфигурации.

```python
# Задаем сетку гиперпараметров
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2', None],
    'solver': ['lbfgs', 'liblinear']
}

# Создаем список всех комбинаций
keys = param_grid.keys()
values = param_grid.values()
combinations = list(product(*values))

print(f"Всего комбинаций для проверки: {len(combinations)}")

# Список для храниения результатов
results = []

# Перебираем параметры
for i, combo in enumerate(combinations):
    params = dict(zip(keys, combo))
    
    # lbfgs поддерживает только l2 и none
    if params['solver'] == 'lbfgs' and params['penalty'] not in ['l2', 'none']:
        continue
    # liblinear поддерживает только l1 и l2
    if params['solver'] == 'liblinear' and params['penalty'] not in ['l1', 'l2']:
        continue
        
    # Создаем пайплайн с текущими параметрами
    pipeline = Pipeline([
        ('prep', new_preprocessor),
        ('selector', FeatureSelector(selected_features)),
        ('classifier', LogisticRegression(
            C=params['C'],
            penalty=params['penalty'],
            solver=params['solver'],
            class_weight='balanced', 
            max_iter=1000,
            random_state=RANDOM_STATE
        ))
    ])
    
    # Запускаем кросс-валидацию
    cv_scores = cross_validate(
        estimator=pipeline,
        X=X_train_val,
        y=y_train_val,
        scoring=['average_precision', 'f1', 'precision', 'recall'],
        cv=cv,
        return_train_score=False
    )
    
    # Сохраняем результаты
    results.append({
        'C': params['C'],
        'penalty': params['penalty'],
        'solver': params['solver'],
        'mean_test_average_precision': cv_scores['test_average_precision'].mean(),
        'std_test_average_precision': cv_scores['test_average_precision'].std(),
        'mean_test_f1': cv_scores['test_f1'].mean(),
        'std_test_f1': cv_scores['test_f1'].std()
    })
    

# Превращаем результаты в таблицу
results_df = pd.DataFrame(results)

# Сортируем по лучшей метрике
results_df = results_df.sort_values('mean_test_average_precision', ascending=False).reset_index(drop=True)
results_df['rank'] = results_df.index + 1
```
